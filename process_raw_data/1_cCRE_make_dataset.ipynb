```python
from pyfaidx import Fasta

genome = Fasta('../../../genome/hg38.fa')
```

```python
seqs = []
for i, row in enumerate(tqdm(df.itertuples(), total=len(df))):
    chr, start, end = row.chr, row.start, row.end
    seq = genome[chr][start:end].seq.upper()
    seqs.append(seq)

df['seq'] = seq
```

```python
accession_df = pd.read_csv('../data/cCRE/accession_metadata.csv')
accession_df
```

```python
df = pd.read_csv('../data/cCRE/GRCh38-cCREs.bed', sep='\t', header=None, names=['chr', 'start', 'end', 'name_1', 'name_2', 'cCRE_type'])
df

for i, row in accession_df.iterrows():
    cell_type = row['cell_type']
    cCRE_accession = row['cCRE']
    cCRE_path = f'../data/cCRE/{cCRE_accession}'
    col_names = ['chr', 'start', 'end', 'name', 'col_1', 'col_2', 'col_3', 'col_4', 'color', 'cCRE_type', 'col_5']
    df_1 = pd.read_csv(cCRE_path, sep='\t', header=None, names=col_names)
    df[f'{cell_type}_cCRE_type'] = df_1['cCRE_type']

df
```

```python
cell_types = ['K562', 'HepG2', 'SK-N-SH', 'HCT116', 'A549']
cols = [f'{c}_cCRE_type' for c in cell_types]
df = df[~(df[cols] == 'Low-DNase').all(axis=1)]
df
```

```python
for i, row in tqdm(df.iterrows(), total=len(df)):
    chr, start, end = row['chr'], row['start'], row['end']
    seq = genome[chr][start:end].seq.upper()
    df.loc[i, 'seq'] = seq

df
```

```python
df
```

```python
df.to_csv('../data/cCRE/cCRE_5_cell_types.csv', index=False)
```
